# **Assignment on Uber dataset**

**Load the dataset**

```python
import pandas as pd
df = pd.read_csv('Uber.csv')
```

**Display basic info about dataset**

```python
df.info()
```

**Check for missing values**

```python
df.isnull().sum()
```

**Drop rows with missing values**

```python
a=df.dropna()
a
```

**fill missing values (propose column with unknown value)**

```python
df['START_DATE*'].fillna(df['START_DATE*'].mode()[0], inplace=True)
df['END_DATE*'].fillna(df['END_DATE*'].mode()[0], inplace=True)
df['CATEGORY*'].fillna(df['CATEGORY*'].mode()[0], inplace=True)
df['START*'].fillna(df['START*'].mode()[0], inplace=True)
df['STOP*'].fillna(df['STOP*'].mode()[0], inplace=True)
df['MILES*'].fillna(df['MILES*'].mean(), inplace=True)
df['PURPOSE*'].fillna(df['PURPOSE*'].mode()[0], inplace=True)
df
```

**Check and remove duplicates**

```python
d=df.duplicated()
df
```

```python
d=df.drop_duplicates()
d
```

**Convert START_DATE and END_DATE to datetime**

```python
df['START_DATE*'] = pd.to_datetime(df['START_DATE*'],errors='coerce')
df.info()
```

```python
df['END_DATE*'] = pd.to_datetime(df['END_DATE*'],errors='coerce')
df.info()
```

**Total number of rides per category:**

```python
df['CATEGORY*'].value_counts()
```

**Total miles traveled for each purpose:**

```python
df.groupby('PURPOSE*')['MILES*'].agg(['sum'])
```

**Average distance for business vs. personal rides:**

```python
average_distances = df.groupby('CATEGORY*')['MILES*'].mean()
average_distances
```

**Add a column for cost estimation (assuming $2 per mile):**

```python
df['COST'] = df['MILES*'] * 2
df
```

**Filter rides longer than 50 miles:**

```python
df[df['MILES*']<50]
```

**Filter by specific purpose (e.g., meetings):**

```python
meeting_data = df[df['PURPOSE*'] == 'Meeting']
meeting_data
```

**What is the total number of business trips versus personal trips?**

```python
df['CATEGORY*'].value_counts()
```

**What percentage of trips are business versus personal?**

```python
total_trips = len(df)
business_trips = len(df[df['CATEGORY*'] == 'Business'])
business_percentage = (business_trips / total_trips) * 100
personal_trips = total_trips - business_trips
personal_percentage = (personal_trips / total_trips) * 100
personal_percentage
```
