# **Assignment on Uber dataset**

**Load the dataset**

```python
import pandas as pd
df=pd.read_csv('Uber.csv')
```

**Display basic info about dataset**

```python

print(df.info())
print(df.shape)
print(df.describe())
print(df.columns)
```

**Check for missing values**

```python
df.isnull().sum()
```

**Drop rows with missing values**

```python
df.dropna()
```

**fill missing values (propose column with unknown value)**

```python
df['PURPOSE*'] = df['PURPOSE*'].replace('unknown', '0') 

df['PURPOSE*'] = df['PURPOSE*'].fillna('0')


df
```

**Check and remove duplicates**

```python
print(df.duplicated())
```

```python
print(df.drop_duplicates())
```

**Convert START_DATE and END_DATE to datetime**

```python
df['START_DATE*'] = pd.to_datetime(df['START_DATE*'], format="%d/%m/%Y %H:%M")
print(df.dtypes)
df.info()
```

```python
df['END_DATE*'] = pd.to_datetime(df['END_DATE*'], format="%d/%m/%Y %H:%M")
print(df.dtypes)
df.info()
```

**Total number of rides per category:**

```python

 df['CATEGORY*'].value_counts()
```

**Total miles traveled for each purpose:**

```python
df.groupby('PURPOSE*')['MILES*'].sum()
```

**Average distance for business vs. personal rides:**

```python
df.groupby('CATEGORY*')['MILES*'].mean()
```

**Add a column for cost estimation (assuming $2 per mile):**

```python
df['COST_ESTIMATION*'] = df['MILES*'] * 2
df
```

**Filter rides longer than 50 miles:**

```python

df['MILES*'] > 50


print(df)

```

**Filter by specific purpose (e.g., meetings):**

```python
df['PURPOSE*']=='Meeting'
df
```

**What is the total number of business trips versus personal trips?**

```python
df['CATEGORY*'].value_counts()
```

**What percentage of trips are business versus personal?**

```python
a=df['CATEGORY*'].value_counts()
b=a.sum()
c=(a.get('Business',0)/b)*100
print(f"{c:.2f}%")
```

```python
a=df['CATEGORY*'].value_counts()
b=a.sum()
c=(a.get('Personal',0)/b)*100
print(f"{c:.2f}%")
```
